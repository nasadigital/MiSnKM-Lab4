```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist('tedges.txt')
```

```python
kc_communities = list(nx.algorithms.community.k_clique_communities(G, 3))
```

```python
community_color = {}
cr_color = 2
for i in kc_communities:
    # print(len(i))
    for j in i:
        community_color[j] = cr_color
    cr_color += 1
color_ar = []
for i in G.nodes():
    if i in community_color:
        color_ar.append(community_color[i])
    else:
        color_ar.append(100)
print(cr_color)
```

```python
G_pos = nx.read_edgelist('tedges.txt')
with open('tedges.txt') as f:
    lines = f.readlines()
    for line in lines:
        ar = line.rstrip().split(' ')
        if ar[0] not in community_color or ar[1] not in community_color:
            if G_pos.has_edge(ar[0], ar[1]):
                G_pos.remove_edge(ar[0], ar[1])
            continue
        if community_color[ar[0]] != community_color[ar[1]]:
            if G_pos.has_edge(ar[0], ar[1]):
                G_pos.remove_edge(ar[0], ar[1])
pos = nx.spring_layout(G_pos)  # compute graph layout
plt.figure(figsize=(10, 10))
plt.axis('off')
nx.draw_networkx_nodes(G, pos, node_size=10, cmap=plt.cm.RdYlBu, node_color=color_ar)
nx.draw_networkx_edges(G, pos, alpha=0.3)
plt.show(G)
```

```python
gn_communities_i = nx.algorithms.community.centrality.girvan_newman(G)
gn_communities = tuple(c for c in next(gn_communities_i))
```

```python
community_color = {}
cr_color = 2
for i in gn_communities:
    # print(len(i))
    for j in i:
        community_color[j] = cr_color
    cr_color += 1
color_ar = []
for i in G.nodes():
    if i in community_color:
        color_ar.append(community_color[i])
    else:
        color_ar.append(15)
print(cr_color)
```
